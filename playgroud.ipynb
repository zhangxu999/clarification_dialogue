```python
with open("option.pkl", "rb") as f:
    option_mapping = pickle.load(f)
```

```python
import numpy as np
import pandas as pd
pd.set_option('display.max_rows', 120)
pd.set_option('display.max_columns', 120)
```

```python
data = []
for key in option_mapping:
    value = option_mapping[key][:10]
    for (w1,s1),(w2,s2) in zip(value[:-1],value[1:]):
        data.append((s1,round(s1-s2,5),(w1,w2),key))

data = sorted(data,reverse=True)
data_arr = np.zeros((101,101),dtype=int)

for score,gap,_,_ in data:
    data_arr[round(score*100),round(gap*100)] += 1
```

```python
pd.DataFrame(data_arr)
```

```python
pd.DataFrame(data_arr.sum(axis=1),columns=['score'])
```

```python
pd.DataFrame(data_arr.sum(axis=0),columns=['gap'])
```
